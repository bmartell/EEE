## Cosa è Jupyter

Il <font color=blue>notebook Jupyter</font> è un'applicazione Web tramite la quale è possibile scrivere del codice e del testo che possono essere di volta in volta modificati ed eseguiti in tempo reale. <font color=blue>Jupyter</font> è gratuito e open-source, usato nella didattica e dagli scienziati per creare e condividere le loro ricerche. Risulta essere usato anche da giornalisti per sviluppare storie e mostrare i loro testi, come ad esempio [Machine bias](https://www.propublica.org/article/machine-bias-risk-assessments-in-criminal-sentencing).

Il <font color=blue>notebook Jupyter</font> combina due componenti:
1. una applicazione Web, ossia uno strumento basato su browser che consente la scrittura interattiva di documenti che combinano codice, testi, e rappresentazione dei risultati;
1. un notebook, ossia un documento in grado di contenere tutto il contenuto visibile nell'applicazione, quali input, output delle elaborazioni, testi, immagini e grafici.

<font color=blue>Jupyter</font> offre un ambiente ideale per caricare i dati, scrivere un codice che analizza i dati e ne esegue la visualizzazione e l'elaborazione, scrivere la documentazione, inserire grafici e tabelle. Inoltre semplica il lavoro di uno scienziato agevolando la creazione di report e la loro condivisione in gruppi di lavoro. 

<font color=blue>Jupyter</font> può essere installato su un qualunque sistema operativo tramite la piattaforma <font color=blue>Anaconda</font>. Quest'ultima semplifica la procedura di download delle librerie e moduli di Python e di altri linguaggi di programmazione come R, C++, Java.

## Caratteristiche dell'applicazione Web

<font color=blue>Jupyter</font> permette:
- di scrivere del codice in apposite celle, evidenziandone la sintassi e l'indentazione, completamento con tab, etc;
- l'esecuzione del codice stesso nella cella e presentazione dei risultati;
- la rappresentazione dei risultati in appositi formati, quali HTML, LaTeX, PNG, MarkDown;
- la possibilità di integrare del testo con il linguaggio di markup Markdown.

## Caratteristiche del notebook

I documenti contengono input e output della sessione interattiva. Le diverse parti possono essere intramezzate da testo non eseguibile che si aggiunge al codice, alle immagini, e così via. Questi documenti sono salvati nel formato JSON con estensione `.ipynb` e possono essere esportati in diversi formati quali HTML, LaTeX, PDF.

<a id='section3.1'></a>
### Celle

Un documento è una sequenza di celle. Una cella è un campo di input testuale multi-riga, il cui contenuto può essere mandato in esecuzione. Le celle possono essere di diverso tipo:
- <font color=blue>celle di codice</font>: queste permettono la scrittura e la modifica del codice nel linguaggio di programmazione selezionato, con evidenziazione della sintassi e completamento con i tab. Quando la cella viene mandata in esecuzione, il codice viene mandato al processo kernel associato con il notebook. Il risultato di tale elaborazione viene rimandato alla cella che lo mostra nella sezione di output della cella stessa. Questo output può contenere testo, grafici e tabelle HTML. 

````python
print('Progetto Cosmic Rays Data Analysis!')
````

- <font color=blue>celle di markdown</font>: queste permettono la scrittura e la modifica del testo, strutturato con il linguaggio di markup Markdown. Quando la cella è eseguita il codice Markdown viene trasformato in testo formattato. In queste celle è possibile usare la notazione LaTeX, trasformata in HTML durante l'esecuzione.

- <font color=blue>celle raw</font>: queste permettono di scrivere output direttamente. La cella non viene mandata in esecuzione dal notebook. Un esempio è dato dalla cella di sopra.

- <font color=blue>celle heading</font>: questo tipo di celle è deprecato e rimpiazzato dalle celle di Markdown.

Le celle sono eseguite con l'opzione `Run cells`, che esegue il contenuto della cella selezionata. Le celle possono essere modificate in loco e fatte rieseguire fino a quando non si ottengono i risultati desiderati.

Le diverse celle che si inseriscono nel notebook sono separate (come se fossero porzioni di codice separate), ma fanno parte dello stesso programma.

````python
giocatori=6
print(giocatori)
````

## Rappresentazione grafica

I grafici possono essere generati e mostrati nella sezione di output delle celle di codice. L'integrazione con la libreria grafica matplotlib è possibile con il comando ```%matplotlib```. 

Questo comando può essere invocato in diversi modi:
- con l'opzione ```inline``` che permette di integrare i grafici direttamente sotto la cella che li ha generati e la loro memorizzazione nel documento notebook;

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot([[0,0],[1,1]], linewidth=4, label='diagonale')
plt.show()
````

- con l'opzione ```notebook``` che permette di integrare i grafici direttamente sotto la cella che li ha generati, permettendo lo zoom, il ridimenzionamento del grafico, il salvataggio del grafico localmente, e la loro memorizzazione nel documento notebook;

````python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.plot([[0,0],[1,1]], linewidth=4, label='diagonale')
plt.show()
````
